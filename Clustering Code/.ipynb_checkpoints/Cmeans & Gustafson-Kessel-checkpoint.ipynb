```python
import numpy as np
import matplotlib.pyplot as plt
import c_means as cm
import gustafson as gus
import timeit
import cv2 as cv
from sklearn.datasets import make_moons
from collections import Counter
```

```python
def plot_points(points, fig=None, loc=(0, 1, 2), title='', method="Cmeans", m=10, n=4):
    if method == "Cmeans":
        algo = cm.FuzzyClustering(max_iter=m, n_clusters=n, method="Cmeans", debug=False)
    elif method == "Gustafson–Kessel":
        algo = cm.FuzzyClustering(max_iter=m, n_clusters=n, method="Gustafson–Kessel", debug=False)
    elif method == "Gustafson–Kessel2":
        algo = gus.Fuzzy_Clustering(max_iter=m, n_clusters=n, method="Gustafson–Kessel")

    clusters, membership = algo.fit(points)
    # membership = np.reshape(membership, (membership.shape[1], membership.shape[0]))
    index = membership.argmax(axis=1)
    if fig:
        fig.add_subplot(loc[0], loc[1], loc[2])
    colmap = {0: 'r', 1: 'g', 2: 'b', 3: 'y', 4: 'tab:pink'}
    for i in range(index.shape[0]):
        plt.scatter(points[i][0], points[i][1], c=colmap[index[i]])
    for i in range(clusters.shape[0]):
        plt.scatter(clusters[i][0], clusters[i][1], c='k')
    plt.title(title)
```

```python
def plot_image(real_image, image_vector, method="Cmeans", m=10, n=4, debug=False):
    # if method == "Kmeans":
    #     algo = km.KMeans()
    algo = None
    if method == "Cmeans":
        algo = cm.FuzzyClustering(max_iter=m, n_clusters=n, method="Cmeans")
    elif method == "Gustafson–Kessel":
        algo = cm.FuzzyClustering(max_iter=m, n_clusters=n, method="Gustafson–Kessel")
    elif method == "Gustafson–Kessel2":
        algo = gus.Fuzzy_Clustering(max_iter=m, n_clusters=n, method="Gustafson–Kessel")
    start = timeit.default_timer()
    clusters, membership = algo.fit(image_vector)
    if clusters.shape[1] > 4:
        clusters = clusters[:, :3]
    if debug:
        print("clusters __________________")
        print(clusters)
    stop = timeit.default_timer()
    if debug:
        print('Time: ', stop - start)
    index = membership.argmax(axis=1)
    # plot the image
    processed_image = np.zeros((image_vector.shape[0], clusters.shape[1]))
    for i in range(index.shape[0]):
        processed_image[i] = clusters[index[i]]
        # chunks[()]processed_image[i]
    processed_image = processed_image.reshape(real_image.shape[0], real_image.shape[1], real_image.shape[2])
    # temp = cv.cvtColor()

    return clusters, processed_image
```

```python
# first data set
X = np.array([12, 20, 28, 18, 29, 33, 24, 45, 45, 52, 51, 52, 53])
Y = np.array([39, 36, 30, 52, 54, 46, 55, 59, 63, 70, 66, 63, 58])
data = np.stack((X, Y), axis=-1)
np.random.seed(100)
```

```python
fig = plt.figure(figsize=(4, 10))
plot_points(data, fig=fig, loc=[2, 1, 1], title="Cmeans", method="Cmeans", m=100, n=4)
plot_points(data, fig=fig, loc=[2, 1, 2], title="Gustafson–Kessel", method="Gustafson–Kessel", m=100, n=4)

plt.show()
```

```python
fig = plt.figure(figsize=(10, 10))
fig.add_subplot(2, 2, 1)
plt.scatter(all_data[:, 0], all_data[:, 1], c='k')
plt.title('Raw data')

plot_points(all_data, fig=fig, loc=[2, 2, 2], title='Cmeans', method="Cmeans", m=20, n=4)
plot_points(all_data, fig=fig, loc=[2, 2, 3], title='Gustafson–Kessel', method="Gustafson–Kessel", m=50, n=4)
plot_points(all_data, fig=fig, loc=[2, 2, 4], title='Gustafson–Kessel2', method="Gustafson–Kessel2", m=20, n=2)
plt.show()
```

```python
# moon shaped data set
moon_data_set = np.array(make_moons(n_samples=500, noise=0.05, random_state=0)[0])
moon_data_set = np.multiply(moon_data_set, 40)
```

```python
fig = plt.figure(figsize=(5, 10))
fig.add_subplot(3, 1, 1)
plt.scatter(moon_data_set[:, 0], moon_data_set[:, 1], c='k')
plt.title('Raw data')

plot_points(moon_data_set, fig=fig, loc=[3, 1, 2], title='Cmeans', method="Cmeans", m=50, n=2)
plot_points(moon_data_set, fig=fig, loc=[3, 1, 3], title='Gustafson–Kessel', method="Gustafson–Kessel", m=50, n=2)
plt.show()
```

```python
image = plt.imread('./example_datasets/gentelman.jpg')
ravel_image = image.reshape(image.shape[0] * image.shape[1], image.shape[2])
if any(ravel_image[i][0] > 1 for i in range(ravel_image.shape[0])):
    print("converting")
    ravel_image = ravel_image / 255.0
```

```python
fig = plt.figure(figsize=(10, 10))
fig.add_subplot(2, 2, 1)
plt.title("Raw image")
plt.imshow(image)

# apply fuzzy to image (Cmeans)
_, clustered_image = plot_image(image, ravel_image, method="Cmeans", m=20, n=4)
clustered_image = np.uint8(clustered_image * 255)
fig.add_subplot(2, 2, 2)
plt.imshow(clustered_image)
plt.title("Cmeans Clustering")

plt.show()
```
